```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from pandas.plotting import andrews_curves, parallel_coordinates, radviz, scatter_matrix
import numpy as np

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target_names[iris.target] # Adds the species name
```

```python
# Set a visually appealing style for Seaborn plots
sns.set_style("whitegrid")
```

```python
# Scatter Plot (Sepal Length vs. Sepal Width, colored by species)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=df)
plt.title('Sepal Length vs. Sepal Width')
plt.show()
```

```python
# Scatter Plot (Petal Length vs. Petal Width, colored by species)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target', data=df)
plt.title('Petal Length vs. Petal Width')
plt.show()
```

```python
# Pair Plot (Relationships between all features)
sns.pairplot(df, hue='target')
plt.title('Pair Plot of All Features')
plt.show()
```

```python
# Box Plot (Sepal Length distribution for each species)
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='sepal length (cm)', data=df)
plt.title('Sepal Length Distribution per Species')
plt.show()
```

```python
# Box Plot (Sepal Width distribution for each species)
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='sepal width (cm)', data=df)
plt.title('Sepal Width Distribution per Species')
plt.show()
```

```python
# Box Plot (Petal Length distribution for each species)
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='petal length (cm)', data=df)
plt.title('Petal Length Distribution per Species')
plt.show()
```

```python
# Box Plot (Petal Width distribution for each species)
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='petal width (cm)', data=df)
plt.title('Petal Width Distribution per Species')
plt.show()
```

```python
# Violin Plot (Petal Length distribution for each species)
plt.figure(figsize=(8, 6))
sns.violinplot(x='target', y='petal length (cm)', data=df)
plt.title('Petal Length Distribution per Species (Violin Plot)')
plt.show()
```

```python
# Histogram (Distribution of Sepal Length)
plt.figure(figsize=(8, 6))
sns.histplot(df['sepal length (cm)'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

```python
# Histogram (Distribution of Petal Length)
plt.figure(figsize=(8, 6))
sns.histplot(df['petal length (cm)'], kde=True)
plt.title('Distribution of Petal Length')
plt.show()
```

```python
# KDE Plot (Distribution of sepal length for each species)
plt.figure(figsize=(8,6))
sns.kdeplot(data=df, x="sepal length (cm)", hue="target", fill=True)
plt.title("Sepal Length Distribution by Species (KDE)")
plt.show()

```

```python
# KDE Plot (Distribution of petal length for each species)
plt.figure(figsize=(8,6))
sns.kdeplot(data=df, x="petal length (cm)", hue="target", fill=True)
plt.title("Petal Length Distribution by Species (KDE)")
plt.show()
```

```python
# Andrews Curves
plt.figure(figsize=(10, 8))
andrews_curves(df, 'target', color=['red', 'green', 'blue'])
plt.title('Andrews Curves for Iris Dataset')
plt.show()
```

```python
# Parallel Coordinates Plot
plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'target', color=['red', 'green', 'blue'])
plt.title('Parallel Coordinates Plot')
plt.show()
```

```python
# RadViz
plt.figure(figsize=(8, 8))
radviz(df, 'target', color=['red', 'green', 'blue'])
plt.title('RadViz for Iris Dataset')
plt.show()
```
