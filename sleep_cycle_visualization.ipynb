```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "sleep_cycle_productivity.csv"
df = pd.read_csv(file_path)

# Display first few rows
df.head()

```

```python
df.info()
```

```python
# Histogram of Sleep Duration
plt.figure(figsize=(8,5))
sns.histplot(df['Total Sleep Hours'], bins=20, kde=True, color='blue')
plt.title('Distribution of Total Sleep Hours')
plt.xlabel('Total Sleep Hours (hours)')
plt.ylabel('Frequency')
plt.show()

```

```python
# Scatter plot of Sleep Duration vs Productivity Score
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Total Sleep Hours'], y=df['Productivity Score'], color='green')
plt.title('Total Sleep Hours vs Productivity Score')
plt.xlabel('Total Sleep Hours (hours)')
plt.ylabel('Productivity Score')
plt.show()
```

```python
# Boxplot of Productivity Score by Sleep Quality
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Sleep Quality'], y=df['Productivity Score'], palette='coolwarm')
plt.title('Productivity Score by Sleep Quality')
plt.xlabel('Sleep Quality')
plt.ylabel('Productivity Score')
plt.show()
```

```python
# Line plot of Sleep Duration over Days
plt.figure(figsize=(10,5))
sns.lineplot(x=df['Date'], y=df['Total Sleep Hours'], marker='o', color='purple')
plt.title('Sleep Duration Over Time')
plt.xlabel('Date')
plt.ylabel('Sleep Duration (hours)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,5))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
