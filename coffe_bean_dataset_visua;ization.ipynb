```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv('DatasetForCoffeeSales2.csv')
df.head()
```

```python
df.info()
```

```python
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Set seaborn style
sns.set(style="whitegrid")
```

```python
# 1. Sales Trend Over Time
plt.figure(figsize=(12, 6))
df.groupby('Date')['Final Sales'].sum().plot()
plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Final Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Top 10 Cities by Sales
plt.figure(figsize=(12, 6))
df.groupby('City')['Final Sales'].sum().nlargest(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 Cities by Sales")
plt.xlabel("City")
plt.ylabel("Total Final Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Sales Distribution by Product Category
plt.figure(figsize=(8, 8))
df.groupby('Category')['Final Sales'].sum().plot(kind='pie', autopct='%1.1f%%', cmap='coolwarm')
plt.title("Sales Distribution by Product Category")
plt.ylabel("")
plt.show()
```

```python
# 4. Top-Selling Products
plt.figure(figsize=(12, 6))
df.groupby('Product')['Final Sales'].sum().nlargest(10).plot(kind='bar', color='lightcoral')
plt.title("Top-Selling Products")
plt.xlabel("Product")
plt.ylabel("Total Final Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Impact of Discounts on Sales
plt.figure(figsize=(8, 6))
sns.boxplot(x='Used_Discount', y='Sales Amount', data=df, palette='coolwarm')
plt.title("Impact of Discounts on Sales")
plt.xlabel("Used Discount")
plt.ylabel("Sales Amount")
plt.show()
```

```python
# 6. Quantity vs. Final Sales (Correlation)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Quantity', y='Final Sales', data=df, alpha=0.7, color='purple')
plt.title("Quantity vs. Final Sales")
plt.xlabel("Quantity")
plt.ylabel("Final Sales")
plt.show()
```

```python
# 7. Distribution of Unit Price
plt.figure(figsize=(8, 6))
sns.histplot(df['Unit Price'], bins=20, kde=True, color='green')
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.show()
```

```python
# 8. Customer Purchase Frequency
plt.figure(figsize=(12, 6))
df['Customer_ID'].value_counts().nlargest(10).plot(kind='bar', color='orange')
plt.title("Top 10 Customers by Purchase Frequency")
plt.xlabel("Customer ID")
plt.ylabel("Number of Purchases")
plt.xticks(rotation=45)
plt.show()
```

```python
# 9. Discount Amount Distribution
plt.figure(figsize=(8, 6))
sns.histplot(df['Discount_Amount'], bins=20, kde=True, color='red')
plt.title("Discount Amount Distribution")
plt.xlabel("Discount Amount")
plt.show()
```
