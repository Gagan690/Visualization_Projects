```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data (you already have this in your Colab)
df = pd.read_csv('sales_data.csv')
```

```python
# Convert 'Sale_Date' to datetime objects
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'])

# Total Sales Over Time
plt.figure(figsize=(12, 6))
df.groupby('Sale_Date')['Quantity_Sold'].sum().plot(kind='line')
plt.title('Total Sales Over Time')
plt.xlabel('Sale Date')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# Sales by Product Category (Bar Chart)
plt.figure(figsize=(10, 6))
df.groupby('Product_Category')['Quantity_Sold'].sum().plot(kind='bar')
plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 3. Distribution of Unit Prices (Histogram)
plt.figure(figsize=(8, 5))
sns.histplot(df['Unit_Price'], kde=True)
plt.title('Distribution of Unit Prices')
plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Quantity Sold vs. Unit Price (Scatter Plot)
plt.figure(figsize=(8, 5))
plt.scatter(df['Unit_Price'], df['Quantity_Sold'])
plt.title('Quantity Sold vs. Unit Price')
plt.xlabel('Unit Price')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# Sales by Product Category (Pie Chart)
plt.figure(figsize=(8, 8))
df.groupby('Product_Category')['Quantity_Sold'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales by Product Category (Proportion)')
plt.ylabel('')
plt.show()
```

```python
# Average Unit Price per Product Category (Bar Chart)
plt.figure(figsize=(10, 6))
df.groupby('Product_Category')['Unit_Price'].mean().plot(kind='bar')
plt.title('Average Unit Price per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Unit Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sales Distribution by Customer Type (Histogram)
plt.figure(figsize=(10, 6))
df.groupby('Customer_Type')['Quantity_Sold'].sum().hist(bins=20)
plt.title('Sales Distribution by Customer Type')
plt.xlabel('Total Quantity Purchased')
plt.ylabel('Number of Customers')
plt.show()

```

```python
# Top 10 Customers by Total Sales (Using Customer_Type as a proxy since Customer_ID is not available)
top_10_customers = df.groupby('Customer_Type')['Quantity_Sold'].sum().nlargest(10) # Using Customer Type
plt.figure(figsize=(10, 6))
top_10_customers.plot(kind='bar')
plt.title('Top Customers by Total Sales (Based on Customer Type)')
plt.xlabel('Customer Type')
plt.ylabel('Total Quantity Sold')
plt.show()
```

```python
# Distribution of Sales by Day of the Week
df['day_of_week'] = df['Sale_Date'].dt.day_name()
plt.figure(figsize=(10,6))
sns.countplot(x='day_of_week', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Distribution of Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Sales')
plt.show()
```
