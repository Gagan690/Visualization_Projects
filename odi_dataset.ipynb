```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("ODI Cricket Data new.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['role'], y=df['total_runs'], estimator=sum, palette='viridis')
plt.title('Total Runs Scored by Player Role')
plt.xlabel('Role')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
df['strike_rate'] = pd.to_numeric(df['strike_rate'], errors='coerce')  # Convert to numeric
sns.histplot(df['strike_rate'], bins=30, kde=True, color='blue')
plt.title('Distribution of Strike Rates')
plt.xlabel('Strike Rate')
plt.ylabel('Number of Players')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['total_wickets_taken'], y=df['total_runs_conceded'], hue=df['role'], alpha=0.7)
plt.title('Wickets Taken vs Runs Conceded')
plt.xlabel('Total Wickets Taken')
plt.ylabel('Total Runs Conceded')
plt.legend(title='Role')
plt.show()
```

```python
plt.figure(figsize=(8,8))
match_results = df[['matches_won', 'matches_lost']].sum()
plt.pie(match_results, labels=['Matches Won', 'Matches Lost'], autopct='%1.1f%%', colors=['green', 'red'])
plt.title('Match Results Distribution')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=df['team'], y=df['player_of_match_awards'], palette='coolwarm')
plt.title('Player of Match Awards by Team')
plt.xlabel('Team')
plt.ylabel('Player of Match Awards')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['total_balls_faced'], y=df['total_runs'], color='purple', marker='o')
plt.title('Runs Scored vs Balls Faced')
plt.xlabel('Total Balls Faced')
plt.ylabel('Total Runs')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Player Performance')
plt.show()
```
