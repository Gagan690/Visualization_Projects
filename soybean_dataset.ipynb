```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = "Advanced Soybean Agricultural Dataset.csv"
df = pd.read_csv(file_path)
```

```python
# Display first few rows
print(df.head())
```

```python
# Check column names and data types
print(df.info())
```

```python
# Histogram of Seed Yield per Unit Area (SYUA)
plt.figure(figsize=(8,5))
sns.histplot(df['Seed Yield per Unit Area (SYUA)'], bins=20, kde=True, color='blue')
plt.title('Distribution of Soybean Seed Yield')
plt.xlabel('Seed Yield per Unit Area (SYUA)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of Seed Yield vs Plant Height
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Plant Height (PH)'], y=df['Seed Yield per Unit Area (SYUA)'], color='green')
plt.title('Soybean Seed Yield vs Plant Height')
plt.xlabel('Plant Height (PH)')
plt.ylabel('Seed Yield per Unit Area (SYUA)')
plt.show()
```

```python
# Line plot of Seed Yield vs Relative Water Content in Leaves
plt.figure(figsize=(10,5))
sns.lineplot(x=df['Relative Water Content in Leaves (RWCL)'], y=df['Seed Yield per Unit Area (SYUA)'], marker='o', color='purple')
plt.title('Soybean Seed Yield vs Relative Water Content in Leaves')
plt.xlabel('Relative Water Content in Leaves (RWCL)')
plt.ylabel('Seed Yield per Unit Area (SYUA)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot of key numerical features
sns.pairplot(df[['Plant Height (PH)', 'Number of Pods (NP)', 'Biological Weight (BW)', 'Seed Yield per Unit Area (SYUA)']], diag_kind='kde')
plt.show()
```

```python
# Violin plot of Protein Percentage vs Seed Yield
plt.figure(figsize=(18,5))
sns.violinplot(x=df['Protein Percentage (PPE)'], y=df['Seed Yield per Unit Area (SYUA)'], palette='muted')
plt.title('Violin Plot of Protein Percentage vs Seed Yield')
plt.xlabel('Protein Percentage (PPE)')
plt.ylabel('Seed Yield per Unit Area (SYUA)')
plt.show()
```

```python
# Density Plot of Seed Yield per Unit Area
plt.figure(figsize=(8,5))
sns.kdeplot(df['Seed Yield per Unit Area (SYUA)'], shade=True, color='blue')
plt.title('Density Plot of Seed Yield per Unit Area')
plt.xlabel('Seed Yield per Unit Area (SYUA)')
plt.ylabel('Density')
plt.show()
```

```python
# Bar plot of Average Seed Yield by Leaf Area Index
plt.figure(figsize=(10,5))
sns.barplot(x=df['Leaf Area Index (LAI)'], y=df['Seed Yield per Unit Area (SYUA)'], estimator=np.mean, palette='coolwarm')
plt.title('Average Seed Yield by Leaf Area Index')
plt.xlabel('Leaf Area Index (LAI)')
plt.ylabel('Average Seed Yield per Unit Area (SYUA)')
plt.xticks(rotation=45)
plt.show()
```
