```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df= pd.read_csv('data.csv')
df.head()
```

```python
print(df.columns.tolist())
```

```python
df.columns = df.columns.str.strip()  # Remove leading and trailing spaces
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Jassids'], bins=15, kde=True, color='blue')
plt.title('Distribution of Jassids Count')
plt.xlabel('Jassids')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['RF'], y=df['Jassids'], color='green')
plt.title('Jassids Count vs Rainfall')
plt.xlabel('Rainfall (RF)')
plt.ylabel('Jassids Count')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df['Wind Velocity'], marker='o', color='red')
plt.title('Wind Velocity Trend Over Time')
plt.xlabel('Index')
plt.ylabel('Wind Velocity')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
sns.pairplot(df, diag_kind='kde', corner=True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x=df['RF'], y=df['Jassids'], palette='coolwarm')
plt.title('Violin Plot of Jassids Count vs Rainfall')
plt.xlabel('Rainfall (RF)')
plt.ylabel('Jassids Count')
plt.show()
```

```python
df['RH-I Range'] = pd.cut(df['RH-I'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
plt.figure(figsize=(8,5))
sns.barplot(x=df['RH-I Range'], y=df['Jassids'], estimator=np.mean, palette='viridis')
plt.title('Average Jassids Count by RH-I Levels')
plt.xlabel('RH-I Range')
plt.ylabel('Average Jassids Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Temp Max'], shade=True, color='red', label='Temp Max')
sns.kdeplot(df['Temp Min'], shade=True, color='blue', label='Temp Min')
plt.title('Density Distribution of Temperature Readings')
plt.xlabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.regplot(x=df['Temp Max'], y=df['Jassids'], color='purple')
plt.title('Regression Plot of Jassids Count vs Temp Max')
plt.xlabel('Temperature Max (°C)')
plt.ylabel('Jassids Count')
plt.show()
```
